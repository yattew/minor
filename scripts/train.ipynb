```python
import json
from pathlib import Path
import gensim
from gensim.models import Word2Vec

from stopstem import *
```

```python
import os
 

path = "C://Users//ganes//Desktop//mini//reference"
dir_list = os.listdir(path)
 
```

```python
print(dir_list)
```

```python
matrix=[]
for i in dir_list:
    with open("../reference/"+i,'r') as data_source:
        data1 = data_source.read()
        data1 = data1.split("\n\n")
        stemmed = list(map(tokenizeAndStem,data1))
        matrix.extend(stemmed)
        
print(matrix)
```

```python
model1 = Word2Vec(matrix, min_count = 1, window = 5, sg=0) 
```

```python
print(model1.wv.similarity(' '.join(tokenizeAndStem('program to reverse an array')),' '.join(tokenizeAndStem("program to print an array"))))
```

```python
import json
from pathlib import Path
import gensim
from gensim.models import Word2Vec

from stopstem import *
import os
 

path = "C://Users//ganes//Desktop//mini//reference"
dir_list = os.listdir(path)
#print(dir_list)
matrix=[]
for i in dir_list:
    with open("../reference/"+i,'r') as data_source:
        try:
            data1 = data_source.read()
            stemmed = tokenizeAndStem(data1)
            matrix.append(stemmed)
        except:
            print(i)
        
#print(matrix)
model1 = Word2Vec(matrix, min_count = 1, window = 5, sg=1) 
```

```python
from cosines import cacCosine
q1 = "Python Program to Print Natural Numbers Using While and For Loop in Hindi - Tutorial #22"
q2 = "Print the first 10 natural numbers using for loop"
print(model1.wv)
def sim(q1,q2):
    a1 = q1.split()
    import numpy as np
    v1 = np.array([0.0]*100)
    v2 = np.array([0.0]*100)
    c1=1
    for i in a1:
        try:
            v1+=model1.wv[i]
            c1+=1
        except:
            pass
    v1/=c1
    a2 = q2.split()
    c2=1
    for i in a2:
        try:
            v2+=model1.wv[i]
            c2+=1
        except:
            pass
    v2/=c2
    return cacCosine(list(v1),list(v2))
import os
BASE_DIR = Path(os.getcwd()).resolve().parent

def findSimilar(listOfDocs,query):
    #dic={stemmdtitle:[title,link,file]}
    dic={}
    maxi=[]
    for i in listOfDocs:
        if i[-4::]=='json':
            with open(BASE_DIR/('jsons/'+i)) as f:
                data=json.load(f)
                for j in data:
                    for k in data[j]['videos'].keys():
                        tok = ' '.join(tokenizeAndStem(k))
                        dic[tok]=[k,data[j]['videos'][k],i]
                        maxi.append([sim(tok,' '.join(tokenizeAndStem(query))),k,i])
                    for k in data[j]['reading'].keys():
                        tok = ' '.join(tokenizeAndStem(k))
                        dic[tok]=[k,data[j]['reading'][k],i]
                        maxi.append([sim(tok,' '.join(tokenizeAndStem(query))),k,i])
    return dic,maxi
```

```python
from pprint import pprint as p
l = os.listdir("C://Users//ganes//Desktop//mini//jsons")
dic,maxi = findSimilar(l,"making dictionary")
p(sorted(maxi,reverse=True))
```
